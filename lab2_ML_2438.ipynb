```python
class uni:
    camp=5#	--- public
    __camp=5#	-- private
    _camp=5#	-- protected
    def __init__(self,name,loc):#----constructor
        self.name=name#	--local name updated
        self.loc=loc
    def display(self):
        print("NAme is",self.name+"location",self.loc)
        print (f"Name is {self.name} and location is {self.loc}")
obj1=uni("FAST","ISB")
obj1.display()
obj1.camp=6
print(obj1.camp)#	-------6
obj2=uni("FAT","IB")
print(obj2.camp)#	-------5
```

```python
class Department(uni):#	--- inheritance
    def __init__(self,name,prog,u_name,u_loc):
        super().__init__(u_name,u_loc)#		----uses child class
        self.prog=prog
    def display(self):
        super().display()
        print("NAme is",self.name+"location",self.prog)
```

```python
obj1=Department("AI","5","G","I")
obj1.display()
```
