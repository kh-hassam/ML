```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
import scipy.stats as stats
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv("Housing.csv")
```

```python
f=[]
cities=['A','B','C','D','E']
```

```python
for i in cities:
    for z in cities:
        for k in cities:
            f.append(i+z+k)
```

```python
count=0
count1=0
for j in f:
    if (j[0]==j[1] or j[1]==j[2] or j[0]==j[2]):
        count=count+1
    if (j[0]!=j[1] and j[1]!=j[2] and j[0]!=j[2]):
        count1+=1
print(count)
count1
```

```python
probability=count/len(f)
probability1=count1/len(f)
p=[]
p.append(probability1)
p.append(probability)
p
```

```python
c=['D','S']
plt.bar(c,p)
```

```python
lab=preprocessing.LabelEncoder()
df['mainroad']=lab.fit_transform(df['mainroad'])
df['guestroom']=lab.fit_transform(df['guestroom'])
df['basement']=lab.fit_transform(df['basement'])
df['hotwaterheating']=lab.fit_transform(df['hotwaterheating'])
df['airconditioning']=lab.fit_transform(df['airconditioning'])
df['prefarea']=lab.fit_transform(df['prefarea'])
df['furnishingstatus']=lab.fit_transform(df['furnishingstatus'])
Y=df.iloc[:,-1]
X=df.drop('furnishingstatus',axis=1)
```

```python
X_train,X_test,Y_train,y_test=train_test_split(X,Y,test_size=0.32,random_state=42)
clf=tree.DecisionTreeClassifier()
clf=clf.fit(X_train,Y_train)
pre=clf.predict(X_test)
accuracy_score(y_test,pre)*100
```
